```python
import requests
import pandas as pd
import json

#url = r'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/New%20York%20City%2CUSA/2023-12-01/2023-12-31?unitGroup=metric&include=hours&key=T2KQ8BWQR28QEPNZKM3DPZMUY&contentType=json'
url = r'https://raw.githubusercontent.com/s-surya-s/IA-Final/main/1_Data/weather_data.txt'

#Make a GET request to the API endpoint
resp = requests.get(url)

#Parse the JSON response
data = json.loads(resp.text)

weather_df = pd.DataFrame()

for day in data['days']:
    
    for data in day['hours']:

        #Fetch key names
        col_name = list(data.keys())

        data['stations'] = ','.join(data['stations'])

        #Create a dataframe with column names (keys)
        temp_df = pd.DataFrame(data, columns=col_name, index=[0])

        temp_df['datetime'] = day['datetime']+' '+temp_df['datetime']

        weather_df = pd.concat([weather_df, temp_df], ignore_index=True)

weather_df.head()
```
